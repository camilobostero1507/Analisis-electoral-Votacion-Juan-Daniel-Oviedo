```python
import pandas as pd
```

```python
df=pd.read_excel(r'C:/Users/KEVIN/Downloads/e22v1bog_nombres (3).xlsx')
```

```python
df_filtrado = df[['Nombre de localidad', 'Votos Oviedo', 'Votos Galán']]
```

```python
df_filtrado['Votos total localidad Oviedo'] = (
    df_filtrado.groupby('Nombre de localidad')['Votos Oviedo']
    .transform('sum')
)

df_filtrado['Votos total localidad Galán'] = (
    df_filtrado.groupby('Nombre de localidad')['Votos Galán']
    .transform('sum')
)
```

```python
df_resumen = (
    df_filtrado
    .groupby('Nombre de localidad', as_index=False)
    .agg({
        'Votos Oviedo': 'sum',
        'Votos Galán': 'sum'
    })
)
```
